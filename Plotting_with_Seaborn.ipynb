# Relational Plots
1. Scatter Plot
2. Line Plot

```python
# Scatter Plots

tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# Axes level function
sns.scatterplot(data=tips, x='total_bill', y = 'tip')
```

```python
# Figure level function
sns.relplot(data=tips, x='total_bill', y = 'tip', kind='scatter', hue='sex', style='time')
```

```python
# Line plot
country = px.data.gapminder()
country
```

```python
India = country[country['country']=='India']
```

```python
India
```

```python
sns.lineplot(data=India, x='year', y='lifeExp')
```

```python
sns.relplot(data=India, x='year', y='lifeExp', kind='line')
```

```python
# facet plot -> work only with figure level function
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', row='sex')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex', row='smoker')
```

```python
# Column Wrap -> No of plots in one row
sns.relplot(data=country, x='gdpPercap', y='lifeExp', col='year', col_wrap=4, kind='scatter')
```

# Matrix Plot
1. Heatmap
2. Clustermap

```python
country
```

```python
country.pivot(index='country', columns='year', values='lifeExp')
```

```python
# heatmap is axes level function and there is no figure level function for this 
plt.figure(figsize=(12,10))
sns.heatmap(country.pivot(index='country', columns='year', values='lifeExp'))
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(country[country['continent']=='Asia'].pivot(index='country', columns='year', values='lifeExp'), annot=True, lineWidth=0.5)
```

```python
plt.figure(figsize=(12,10))

sns.heatmap(country[country['continent']=='Asia'].pivot(index='country', columns='year', values='lifeExp'), annot=True, lineWidth=0.5, cmap='viridis')
```

# Categorical Plots

Categorical Scatter plots:-
 1. Stripplot
 2. Swarmplot

```python
iris = sns.load_dataset('iris')
```

```python
# strip plot - 1 num col and 1 cat col and will plot Scatter plot
# Axes level function
sns.stripplot(data=tips, x='day', y='total_bill')
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', jitter=False)
```

```python
# catplot -> figure level function

sns.catplot(data=tips, x='day', y='total_bill', kind='strip')
```

```python
# swarmplot

sns.catplot(data=tips, x='day', y='total_bill', kind='swarm')
```

```python
# axes level
sns.swarmplot(data=tips, x='day', y='total_bill')
```

# MultiGridPlots

1. facetgrid
2. pairgrid
3. joingrid

```python
# FacetGrid
g = sns.FacetGrid(data=tips, col='day', row='time')
```

```python
g = sns.FacetGrid(data=tips, col='day', row='time')
g.map(sns.violinplot, 'sex', 'total_bill')
```

```python
# pairplot

sns.pairplot(iris)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
# PairGrid

g = sns.PairGrid(data=iris)

g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris)

g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# Joinplot

sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter')
```
