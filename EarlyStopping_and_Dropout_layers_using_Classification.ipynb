```python
X = df.drop('benign_0__mal_1', axis=1).values
y = df['benign_0__mal_1'].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=48)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=800, validation_data=(X_test, y_test))
```

```python
logs = pd.DataFrame(model.history.history)
```

```python
logs
```

```python
# Model is overfitted as validation loss keep on increasing
logs.plot()
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

EarlyStopping

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=20)
```

```python
model.fit(X_train, y_train, epochs=800, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
logs_after_earlyStopping = pd.DataFrame(model.history.history)
```

```python
logs_after_earlyStopping.plot()
```

Adding Dropout layers

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dropout(rate=0.4)) # 40% of neuron will be turned randomly during training
model.add(Dense(16, activation='relu'))
model.add(Dropout(rate=0.4))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=800, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
logs_after_dropouts = pd.DataFrame(model.history.history)
```

```python
logs_after_dropouts.plot()
```

```python
prediction = (model.predict(X_test)>0.5).astype("int32")
```

```python
prediction
```

```python
y_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, prediction))
print(confusion_matrix(y_test, prediction))
```
