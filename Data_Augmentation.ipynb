Using ImageDataGenerator

```python
# Parameters defined here will tell the type of transformations we need to apply.
dataGenerator = ImageDataGenerator(
    rotation_range=40,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip=True,
    vertical_flip = False,
    width_shift_range = 0.2,
    height_shift_range = 0.2
)
```

```python
# change image type to array

image = keras.utils.img_to_array(image)
```

```python
type(image)
```

```python
image.shape
```

```python
input_batch = image.reshape(1,200,200,3) # 1 is the batch size i.e the no of images we are getting as input.
```

```python
count = 0
# Total 10 new images created after the transformation based on parameters
for images in dataGenerator.flow(input_batch, batch_size=1, save_to_dir='/content/Data'):
  count+=1
  if count==10:
    break
```
