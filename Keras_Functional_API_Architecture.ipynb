```python
from keras.models import Model
```

```python
from keras.layers import *
```

```python
input = Input(shape=(10,))
```

```python
# hidden1 layer is getting input from input layer
hidden1 = Dense(28, activation='relu')(input)
# hidden2 layer is getting input from hidden1 layer
hidden2 = Dense(64, activation='relu')(hidden1)

# output1 and output2 layer is getting input from hidden2 layer
output1 = Dense(1, activation='linear')(hidden2) # Regression predcition
output2 = Dense(1, activation='sigmoid')(hidden2) # Classification prediction
```

```python
model = Model(inputs = input, outputs= [output1, output2])
```

```python
model.summary()
```

```python
from keras.utils import plot_model

plot_model(model, show_shapes=True)
```
