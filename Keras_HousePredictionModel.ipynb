```python
df.drop('id', axis=1, inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['month'] = pd.to_datetime(df['date'], format='%m%d%y').dt.month
```

```python
df['year'] = df['date'].apply(lambda date: date.year)
```

```python
df['day'] = pd.to_datetime(df['date'], format='%m%d%y').dt.day
```

```python
df.drop('date', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['year'].value_counts()
```

```python
df.drop('zipcode',axis=1, inplace=True)
```

```python
X = df.drop('price', axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=48)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='Adam', loss='mse')
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=200, batch_size=64)
```

```python
history_logs = pd.DataFrame(model.history.history)
```

```python
history_logs
```

```python
history_logs.plot()
```

```python
 predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predictions))
```

```python
rmse
```

```python
explained_variance_score(y_test, predictions)
```
